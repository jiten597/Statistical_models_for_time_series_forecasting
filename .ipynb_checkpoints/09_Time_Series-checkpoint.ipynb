## Defining SARIMA model
$SARIMA (p,w,q)(P,W,Q)_m$ adds another set of parameters that allows us to take into account periodic patterns when forecasting a time series which is not always possible with an ARIMA(p,w,q) model.

Parameters to the model: P, W, Q and m. The first three have the same meaning as in the ARIMA(p,w,q) model, only they are their seasonal counterparts. To understand the meaning of these parameters and how they affect the final model, we must first define the parameter m. The parameter m stands for the frequency. In the context of time series, the frequency is defined as the number of observations per cycle. `P` is the order of the seasonal AR(P) process, `W` is the seasonal order of integration, `Q` is the order of the seasonal MA(Q) process, and m is the frequency or the number of observations per seasonal cycle. 

If the data is recorded annually, then m = 1, since there is only one observation per year. If the data is recorded quarterly, then m = 4, since there are four quarters in a year. If the data is recorded monthly, then m = 12. Finally, for weekly data, m = 52.

**Appropriate frequency (m) depending on the data:**

| Data | Frequency (m) |
| --- | --- |
| Anuualy | 1 |
| Quarterly | 4 |
| Monthly | 12 |
| Weekly | 52 |

**Appropriate frequency (m) for daily and sub-daily data:**

| Data | Minute | Hour | Day | Week | Year |
| --- | --- | --- | --- | --- | --- |
| Daily |  |  |  | 7 | 365 |
| Hourly |  |  | 24 | 168 | 8766 |
| Every minute |  | 60 | 1440 | 10080 | 525960 |
| Every second | 60 | 3600 | 86400 | 604800 | 31557600 |

**Example:**
Let m = 12. If P = 2 means that we `include two past values` of the series at lag that is a multiple of m. Therefore, we include the value at $y_{t-12}$ and $y_{t-24}$. Similarly, if D = 1, then it means that a seasonal difference makes the series stationary. In this case, a seasonal difference would be expressed as

$ùë¶'_t = ùë¶_ùë° ‚àí ùë¶_{ùë°‚àí12}$

Then, in a situation where Q = 2, this means that we include past error terms at lags that are a multiple of m. Therefore, we include the error $Œµ_{t-12}$ and $Œµ_{t-24}$.



## Identifying seasonal patterns in a time series
* plotting the time series data
* time series decomposition

### Time series decomposition
It is a statistical task that separates the time series into its three main components: `trend`, `seasonal`, and the `residuals`. **It is a graphical method and not a statistical test.**

**The trend component represents the long-term change in the time series. This component is responsible for time series that increase or decrease over time. The seasonal component is of course the periodic pattern in the time series. It represents repeated fluctuations that occur over a fixed period of time. Finally, the residuals, or the noise, express any irregularity that cannot be explained by the trend or the seasonal component.**

## Forecasting
### Forecasting with an ARIMA model 

```python
## Test for stationarity
def adf_test(series):
    result = adfuller(series)
    print('ADF Statistics: {}'.format(result[0])) # Retrieve the ADF statistic
    print('p - value: {}'.format(result[1]))      # Retrieve the p-value
    if result[1] <= 0.05:
        print("strong evidence against the H0, reject the H0. Data has no unit root and is stationary")
    else:
        print("weak evidence against H0, time series has a unit root, indicating it is Non-stationary ")
    
    return result

ADF_result = adf_test(df['Thousands of Passengers'])
```

```python
df_diff    = np.diff(df['Thousands of Passengers'], n=1)
ADF_result = adf_test(df_diff)
```

```python
eps_diff2  = np.diff(df_diff, n=1) # Series is now differenced twice
ADF_result = adf_test(eps_diff2)
```

Since the data is sampled monthly and we know it is seasonal, we can hypothesize that the number of air passengers in January of a given year is likely predictive of the number of air passengers in January of the following year. Since these `two points are 12 timesteps apart`, we allow the values of p and q to vary between 0 and 12 in order to potentially capture this seasonal information in the ARIMA(p,d,q) model.

```python
ps = range(0, 13, 1)
qs = range(0, 13, 1) 
Ps = [0]  ## Set P and Q to 0, since we are working with an ARIMA(p,w,q) model
Qs = [0] 
w  = 2    ## number of times the series was differenced to become stationary
W  = 0    ## Set W to 0 since we are working with an ARIMA(p,w,q) model
m  = 12   ## frequency for seasonal pattern

ARIMA_order_list = list(product(ps, qs, Ps, Qs))
ARIMA_order_list
```

```python
def optimize_SARIMA(endog: Union[pd.Series, list], order_list: list, w: int, W: int, m: int) -> pd.DataFrame:
    
    results = []
    
    for order in tqdm_notebook(order_list):
        try: 
            model = SARIMAX(
                endog, 
                order               = (order[0], w, order[1]),
                seasonal_order      = (order[2], W, order[3], m),
                simple_differencing = False).fit(disp=False)
        except:
            continue
            
        aic = model.aic
        results.append([order, aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q,P,Q)', 'AIC']
    
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df


train = df['Thousands of Passengers'][:-12]

ARIMA_result_df = optimize_SARIMA(train, ARIMA_order_list, w, W, m)
ARIMA_result_df
```

```python
ARIMA_model     = SARIMAX(train, order=(11,2,3), simple_differencing=False) # SARIMA(11,2,3)(0,0,0)_12 = ARIMA(11,2,3)
ARIMA_model_fit = ARIMA_model.fit(disp=False) 
ARIMA_model_fit.plot_diagnostics(figsize=(12,8))
```

* The top left plot, we can see that the residuals have no trend with a variance that seems fairly constant over time, which resembles the behavior of white noise.
* The top right plot shows the distribution of the residuals which approaches a normal distribution.
* The bottom left plot shows Q-Q plot which displays a fairly straight line that lies on y = x.
* The bottom right plot shows no significant autocorrelation coefficients after lag 0, which is exactly like white noise. From this analysis, the residuals resemble white noise. 

```python
residuals       = ARIMA_model_fit.resid 
lbvalue, pvalue = acorr_ljungbox(residuals, np.arange(1, 11, 1)) 
print(pvalue)

## p-values are all greater than 0.05. Therefore, we do not reject the null hypothesis, meaning that
## the residuals are indeed uncorrelated, just like white noise.
## Therefore, the model passes all tests from the residuals analysis, and it can be used for forecasting. 
```

```python
test                   = df.iloc[-12:]
test['naive_seasonal'] = df['Thousands of Passengers'].iloc[120:132].values
ARIMA_pred             = ARIMA_model_fit.get_prediction(132, 143).predicted_mean ## Get predictions for each month of 1960
test['ARIMA_pred']     = ARIMA_pred 
test
```

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(df['Month'], df['Thousands of Passengers'])
ax.plot(test['Thousands of Passengers'], 'b-',  label='actual')
ax.plot(test['naive_seasonal'],          'r:',  label='naive seasonal')
ax.plot(test['ARIMA_pred'],              'k--', label='ARIMA(11,2,3)')

ax.set_xlabel('Date')
ax.set_ylabel('Number of air passengers')
ax.axvspan(132, 143, color='#808080', alpha=0.2)

ax.legend(loc=2)

plt.xticks(np.arange(0, 145, 12), np.arange(1949, 1962, 1))
ax.set_xlim(120, 143)

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

### Forecasting with a SARIMA model

```python
eps_diff   = np.diff(df['Thousands of Passengers'], n=1) # Series is now differenced once
ADF_result = adf_test(eps_diff)
# eps_diff2  = np.diff(df_diff, n=1) # Series is now differenced twice
# ADF_result = adf_test(eps_diff2)

## Let‚Äôs apply a seasonal difference and test for stationarity.
eps_diff_seasonal_diff = np.diff(eps_diff, n=12)
ADF_result             = adf_test(eps_diff_seasonal_diff)
```

```python
ps = range(0, 4, 1)
qs = range(0, 4, 1)
Ps = range(0, 4, 1)
Qs = range(0, 4, 1)

SARIMA_order_list = list(product(ps, qs, Ps, Qs))

train = df['Thousands of Passengers'][:-12]

w = 1   ## one round of differencing
W = 1   ## one round of seasonal differencing
m = 12

SARIMA_result_df = optimize_SARIMA(train, SARIMA_order_list, w, W, m)
SARIMA_result_df
```

```python
SARIMA_model     = SARIMAX(train, order=(2,1,1), seasonal_order=(1,1,2,12), simple_differencing=False)
SARIMA_model_fit = SARIMA_model.fit(disp=False)

print(SARIMA_model_fit.summary())
```

```python
SARIMA_model_fit.plot_diagnostics(figsize=(12,8))
```

```python
residuals       = SARIMA_model_fit.resid 
lbvalue, pvalue = acorr_ljungbox(residuals, np.arange(1, 11, 1)) 
print(pvalue) 

## The returned p-values are all greater than 0.05. Therefore, we do not reject the null hypothesis and conclude that 
## the residuals are independent and uncorrelated, just like white noise. 
```

```python
SARIMA_pred         = SARIMA_model_fit.get_prediction(132, 143).predicted_mean
test['SARIMA_pred'] = SARIMA_pred
test
```

```python
fig, ax = plt.subplots(figsize=(12,5))

ax.plot(df['Month'], df['Thousands of Passengers'])
ax.plot(test['Thousands of Passengers'], 'b-',  label='actual')
ax.plot(test['naive_seasonal'],          'r:',  label='naive seasonal')
ax.plot(test['ARIMA_pred'],              'k--', label='ARIMA(11,2,3)')
ax.plot(test['SARIMA_pred'],             'g-.', label='SARIMA(2,1,1)(1,1,2,12)')

ax.set_xlabel('Date')
ax.set_ylabel('Number of air passengers')
ax.axvspan(132, 143, color='#808080', alpha=0.2)

ax.legend(loc=2)

plt.xticks(np.arange(0, 145, 12), np.arange(1949, 1962, 1))
ax.set_xlim(120, 143)

fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

## Performance

```python
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


mape_naive_seasonal = mape(test['Thousands of Passengers'], test['naive_seasonal'])
mape_ARIMA          = mape(test['Thousands of Passengers'], test['ARIMA_pred'])
mape_SARIMA         = mape(test['Thousands of Passengers'], test['SARIMA_pred'])

print(mape_naive_seasonal, mape_ARIMA, mape_SARIMA)
```

```python
fig, ax = plt.subplots(figsize=(12,5))

x = ['naive seasonal', 'ARIMA(11,2,3)', 'SARIMA(2,1,1)(1,1,2,12)']
y = [mape_naive_seasonal, mape_ARIMA, mape_SARIMA]

ax.bar(x, y, width=0.4)
ax.set_xlabel('Models')
ax.set_ylabel('MAPE (%)')
ax.set_ylim(0, 15)

for index, value in enumerate(y):
    plt.text(x=index, y=value + 1, s=str(round(value,2)), ha='center')

plt.tight_layout()
plt.show()
```
